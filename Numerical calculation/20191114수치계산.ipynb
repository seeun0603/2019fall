```
#과제 5
from matplotlib import pyplot as plt
import numpy as np

plt.xlabel("x label") #x축 라벨
plt.ylabel("y label") #y축 라벨
plt.title("title") #title

x = np.linspace(0, 2*np.pi, 50)

y = np.sin(x)
y1 = np.sin(2*x)

plt.plot(x,y,'--r')
plt.annotate('sin(x)',xy= (2,-0.75),xytext=(1,-0.75),arrowprops={'color':'green'})
plt.plot(x,y1)
plt.annotate('sin(2x)',xy= (2,0.8),xytext=(0,0),arrowprops={'color':'yellow'})
#xy를 나타내는 text의 위치이고 xy는 가르키는 포인트의 위치

plt.show()
```

```
#과제 6
from matplotlib import pyplot as plt
import numpy as np

plt.xlabel("x label") #x축 라벨
plt.ylabel("y label") #y축 라벨
plt.title("title") #title

x = np.linspace(0, 2*np.pi, 50)

y = np.cos(x)
y1 = 2*np.sin(x)+1

plt.plot(x,y,'--r')
plt.annotate('cos(x)',xy= (2,-0.75),xytext=(1,-0.75),arrowprops={'color':'green'})
plt.plot(x,y1)
plt.annotate('2*np.sin(x)+1',xy= (1.8,2.9),xytext=(2,1),arrowprops={'color':'yellow'})

plt.show()
```
